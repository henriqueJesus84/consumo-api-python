```python
import requests
from datetime import datetime

while True:
    data_inicio = input("Digite a data inicial (YYYY-MM-DD): ")
    data_final = input("Digite a data final (YYYY-MM-DD): ")

    data_inicio_dt = datetime.strptime(data_inicio, "%Y-%m-%d")
    data_final_dt = datetime.strptime(data_final, "%Y-%m-%d")

    diferenca = (data_final_dt - data_inicio_dt).days

    if diferenca < 0:
        print("A data final deve ser maior que a inicial.")

    elif diferenca > 7:
        print("A API da NASA permite no máximo 7 dias.")

    else:
        break


api_key = "DEMO_KEY"

url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date={data_inicio}&end_date={data_final}&api_key={api_key}"

resposta = requests.get(url)
dados = resposta.json()

asteroides = {}

for data in dados["near_earth_objects"]:
    for obj in dados["near_earth_objects"][data]:

        nome = obj["name"]

        asteroides[nome] = {
            "diametro": obj["estimated_diameter"]["meters"]["estimated_diameter_max"],
            "distancia": obj["close_approach_data"][0]["miss_distance"]["kilometers"],
            "perigoso": obj["is_potentially_hazardous_asteroid"],
            "orbita": obj["close_approach_data"][0]["orbiting_body"]
        }

print("\nTotal de asteroides encontrados:", len(asteroides))
print()

for nome, info in asteroides.items():

    perigo = "Sim" if info["perigoso"] else "Não"

    print("Objeto:", nome)
    print("Diâmetro máximo (m):", round(info["diametro"], 2))
    print("Distância mínima da Terra (km):", round(float(info["distancia"]), 2))
    print("Potencialmente perigoso:", perigo)
    print("Orbita:", info["orbita"])
    print("----------------------")
```
